```python
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
path = "../data/Test_knight.csv"
data_test = pd.read_csv(path)

columns = data_test.columns
N = len(columns)
```

```python
path = "../data/Train_knight.csv"
data_train= pd.read_csv(path)
```

```python
col_test = 'Evade'
```

```python
fig, axs = plt.subplots(6, 5, figsize=(12, 12))
for idx, column in enumerate(columns[:N]):
    row = idx // 5
    col = idx % 5
    axs[row, col].scatter(data_test[col_test],data_test[column], marker='o', s=10, alpha=0.5, c='g')
    axs[row, col].set_title(column)
    axs[row, col].legend(['Knight'])
plt.tight_layout()
plt.show()
```

```python
data_sith = data_train[data_train['knight'] == 'Sith']
data_jedi = data_train[data_train['knight'] == 'Jedi']
```

```python
fig, axs = plt.subplots(6, 5, figsize=(12, 12))
for idx, column in enumerate(columns[:N]):
    row = idx // 5
    col = idx % 5
    axs[row, col].scatter(data_sith[col_test],data_sith[column], marker='o', s=10, alpha=0.5, c='r')
    axs[row, col].scatter(data_jedi[col_test],data_jedi[column], marker='o', s=10, alpha=0.5, c='b')
    axs[row, col].set_title(column)
    axs[row, col].legend(['Sith', 'Jedi'])
plt.tight_layout()
plt.show()
```

```python
col_test = 'Evade'

fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].scatter(data_sith[col_test],data_sith['Recovery'], marker='o', alpha=0.5, c='r')
axs[0, 0].scatter(data_jedi[col_test],data_jedi['Recovery'], marker='o', alpha=0.5, c='b')
axs[0, 0].set_title('Recovery')
axs[0, 0].legend(['Sith', 'Jedi'])

axs[0, 1].scatter(data_test[col_test],data_test['Recovery'], marker='o', alpha=0.5, c='g')
axs[0, 1].set_title('Recovery')
axs[0, 1].legend(['Knight'])

axs[1, 0].scatter(data_sith[col_test],data_sith['Agility'], marker='o', alpha=0.5, c='r')
axs[1, 0].scatter(data_jedi[col_test],data_jedi['Agility'], marker='o', alpha=0.5, c='b')
axs[1, 0].set_title('Agility')
axs[1, 0].legend(['Sith', 'Jedi'])


axs[1, 1].scatter(data_test[col_test],data_test['Agility'], marker='o', alpha=0.5, c='g')
axs[1, 1].set_title('Agility')
axs[1, 1].legend(['Knight'])

plt.tight_layout()
plt.show()
```
