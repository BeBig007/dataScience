```python
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
path = "../data/Test_knight.csv"
data_test = pd.read_csv(path)

data_test.head()
```

```python
columns = data_test.columns
N = len(columns)
```

```python
fig, axs = plt.subplots(6, 5, figsize=(15, 15))
for idx, column in enumerate(columns[:N]):
    row = idx // 5
    col = idx % 5
    axs[row, col].hist(data_test[column], bins=40, color='#7fbf7f')
    axs[row, col].set_title(column)
    axs[row, col].legend(['Knight'])
plt.tight_layout()
plt.show()
```

```python
path = "../data/Train_knight.csv"
data_train= pd.read_csv(path)

data_train.head()
```

```python
data_sith = data_train[data_train['knight'] == 'Sith']
```

```python
data_jedi = data_train[data_train['knight'] == 'Jedi']
```

```python

fig, axs = plt.subplots(6, 5, figsize=(15, 15))
for idx, column in enumerate(columns[:N]):
    row = idx // 5
    col = idx % 5
    axs[row, col].hist(data_sith[column], bins=40, color='red', alpha=0.5)
    axs[row, col].hist(data_jedi[column], bins=40, color='blue', alpha=0.5)
    axs[row, col].set_title(column)
    axs[row, col].legend(['Sith', 'Jedi'])
plt.tight_layout()
plt.show()
```
